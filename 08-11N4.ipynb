```python
import pandas as pd
import matplotlib.pyplot as plt

# è¯»å–CSVæ–‡ä»¶
data = pd.read_csv('N4DLC_resnet50_NAug10shuffle1_129000.csv')

# æå–åæ ‡æ•°æ®
leftleg2_x = data['leftleg2_x']
leftleg2_y = data['leftleg2_y']

rightleg2_x = data['rightleg2_x']
rightleg2_y = data['rightleg2_y']

# è®¡ç®—é€Ÿåº¦å‡½æ•°
def calculate_speed(x_values, y_values):
    speeds = []
    for i in range(1, len(x_values)):
        # è®¡ç®—ä½ç½®å˜åŒ–
        dx = x_values[i] - x_values[i - 1]
        dy = y_values[i] - y_values[i - 1]
        
        # è®¡ç®—æ¬§æ°è·ç¦»ä½œä¸ºä½ç§»
        displacement = (dx**2 + dy**2)**0.5
        
        # å‡è®¾æ—¶é—´é—´éš”ä¸º1ï¼ˆå¯ä»¥æ ¹æ®å®žé™…æƒ…å†µè°ƒæ•´ï¼‰
        time_interval = 1
        
        # è®¡ç®—é€Ÿåº¦å¹¶æ·»åŠ åˆ°é€Ÿåº¦åˆ—è¡¨
        speed = displacement / time_interval
        speeds.append(speed)
    
    return speeds

# è®¡ç®—å·¦è…¿é€Ÿåº¦
left_leg_speeds = calculate_speed(leftleg2_x, leftleg2_y)

# è®¡ç®—å³è…¿é€Ÿåº¦
right_leg_speeds = calculate_speed(rightleg2_x, rightleg2_y)

# ç»˜åˆ¶é€Ÿåº¦å›¾
plt.figure(figsize=(10, 6))

plt.plot(left_leg_speeds, color='green', label='Left Leg Speed')
plt.plot(right_leg_speeds, color='purple', label='Right Leg Speed')

plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Fruit Fly Leg Speeds')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# è¯»å–CSVæ–‡ä»¶
data = pd.read_csv('N4DLC_resnet50_NAug10shuffle1_129000.csv')

# æå–åæ ‡æ•°æ®
leftleg2_x = data['leftleg2_x']
leftleg2_y = data['leftleg2_y']

rightleg2_x = data['rightleg2_x']
rightleg2_y = data['rightleg2_y']

# è®¡ç®—é€Ÿåº¦å‡½æ•°
def calculate_speed(x_values, y_values):
    speeds = []
    max_speed = 0.0  # åˆå§‹åŒ–æœ€å¤§é€Ÿåº¦
    
    for i in range(1, len(x_values)):
        # è®¡ç®—ä½ç½®å˜åŒ–
        dx = x_values[i] - x_values[i - 1]
        dy = y_values[i] - y_values[i - 1]
        
        # è®¡ç®—æ¬§æ°è·ç¦»ä½œä¸ºä½ç§»
        displacement = (dx**2 + dy**2)**0.5
        
        # å‡è®¾æ—¶é—´é—´éš”ä¸º1ï¼ˆå¯ä»¥æ ¹æ®å®žé™…æƒ…å†µè°ƒæ•´ï¼‰
        time_interval = 1
        
        # è®¡ç®—é€Ÿåº¦
        speed = displacement / time_interval
        
        # æ›´æ–°æœ€å¤§é€Ÿåº¦
        if speed > max_speed:
            max_speed = speed
        
        speeds.append(speed)
    
    return max_speed, speeds

# è®¡ç®—å·¦è…¿é€Ÿåº¦
max_left_leg_speed, left_leg_speeds = calculate_speed(leftleg2_x, leftleg2_y)

# è®¡ç®—å³è…¿é€Ÿåº¦
max_right_leg_speed, right_leg_speeds = calculate_speed(rightleg2_x, rightleg2_y)

# è¾“å‡ºæ¯åªè…¿çš„æœ€å¤§é€Ÿåº¦
print(f"maxleft_leg2_speed: {max_left_leg_speed}px/s")
print(f"maxright_leg2_speed: {max_right_leg_speed}px/s")

# ç»˜åˆ¶é€Ÿåº¦å›¾ï¼ˆåŒä¹‹å‰çš„ç¤ºä¾‹ï¼‰
plt.figure(figsize=(10, 6))

plt.plot(left_leg_speeds, color='green', label='Left Leg Speed')
plt.plot(right_leg_speeds, color='purple', label='Right Leg Speed')

plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Fruit Fly Leg Speeds')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# è¯»å–CSVæ–‡ä»¶
data = pd.read_csv('N4DLC_resnet50_NAug10shuffle1_129000.csv')

# æå–åæ ‡æ•°æ®
leftleg2_x = data['leftleg2_x']
leftleg2_y = data['leftleg2_y']

# è®¡ç®—æ€»è·¯ç¨‹
total_distance = 0
for i in range(1, len(leftleg2_x)):
    dx = leftleg2_x[i] - leftleg2_x[i - 1]
    dy = leftleg2_y[i] - leftleg2_y[i - 1]
    total_distance += (dx**2 + dy**2)**0.5


# è®¡ç®—æ€»æ—¶é—´
total_time = 3000

# è®¡ç®—é€Ÿåº¦
average_speed = total_distance / total_time

print(f"Total Distance: {total_distance}px")
print(f"Total Time: {total_time}s")
print(f"Average Speed: {average_speed}px/s")
```
